```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
#tugas = prediksikan net income pada 2020
```

```python
df = pd.read_csv('/content/belajarML/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color = 'green', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)']) #fitting the data artinya kita men training model nya berdasar poin data nya
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#rumus prediksi harga : price = m(coef_) * area(year) + b(intercept_)
prediksi = 828.46507522*2020+-1632210.7578554575 #akan menghasilkan hasil dari reg.predict([[2020]])
hasil = 'prediksi 2020 = '
print(hasil, prediksi)
```
